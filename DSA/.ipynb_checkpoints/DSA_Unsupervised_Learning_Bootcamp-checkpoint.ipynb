```python
# Looking for outliers
#visualize outliers for specific column
def outlier_visualizer(col, vert=True, grid=True):
    data[col].plot(kind='box', vert=vert)
    plt.title(f"{col} - Boxplot")
    plt.ylabel('Values')
    plt.grid(grid)
    plt.show()

# identify outliers for a specific column using IQR method
def outlier_finder(col):
    Q1 = np.percentile(data[col],25)
    Q3 = np.percentile(data[col], 75)
    IQR = Q3- Q1
    left_lim = Q1 - 1.5 * IQR
    right_lim = Q3 + 1.5 * IQR
    outliers = { 'country' : [], col:[]}

    for i,c in enumerate(data[col]):
        if c< left_lim:
            outliers['country'].append(data[col].index[i])
            outliers[col].append(c)
        elif c > right_lim:
            outliers['country'].append(data[col].index[i])
            outliers[col].append(c)

    outliers_df = pd.DataFrame(outliers)
    outliers_df.set_index('country', inplace=True)
    print(f"{col} % of outliers : {round(outliers_df.shape[0]/data.shape[0] * 100,2)}%")
    
    return outliers_df

    
        
```

```python
# Making the correlation matrix with strong correaltion only
def corr_matrix(data):
    corr_matrix = data.corr()
    for column in corr_matrix.columns:
        corr_matrix[column] = corr_matrix[column].apply(lambda x: x if abs(x) >0.70 and abs(x) < 1 else '')

    return corr_matrix
```

## Feature Engineering

**Transformation:** It is used to convert the skewed data into normal distribution.

```python
# make a deep copy of the original df and do logaritmic transformation
from scipy.stats import boxcox
```

```python
# Create a function to apply several transformations
def transformer(data,function):
    lam = []
    for col in data.columns:
        if abs(data[col].skew()) > 0.75:
            if function == boxcox:
                data[col], lam_val = function(data[col] + abs(data.min().min())+0.001)
                lam.append(lam_val)
            else:
                data[col] = function(data[col] + abs(data.min().min())+0.001)
                # inflation is the only feature with negative values; to make the log transform possible,
                ## I added the lowest value of inflation +0.001 to make all values positive
    if function == boxcox:
        return data, lam
    else:
        return data
```

```python
log_data = data.copy(deep=True)
log_data = transformer(log_data, np.log)
log_data.skew()
```

```python
sns.pairplot(log_data);
```

```python
corr_matrix(log_data)
```
